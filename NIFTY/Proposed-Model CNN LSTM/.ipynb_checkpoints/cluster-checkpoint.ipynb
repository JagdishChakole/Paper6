```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv('tempcnn1lstm.csv')
```

```python
# Select rows from 60 onward
data_subset = X.iloc[60:].copy()  # 
```

```python
# Apply K-Means on the selected rows
kmeans = KMeans(n_clusters=8, random_state=42)
data_subset['Cluster'] = kmeans.fit_predict(data_subset[['state2d', 'state1d']])
```

```python
# Store the clustered values back in the original DataFrame
X.loc[60:, 'Cluster'] = data_subset['Cluster'].values
```

```python
#df.index = pd.to_datetime(df.index)
X = X.reset_index(drop=True)
X.to_csv('tempcnnlstmcluster.csv', index=True) 
```
