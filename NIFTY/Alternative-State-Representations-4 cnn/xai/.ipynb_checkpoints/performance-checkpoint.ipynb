```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn
import numpy as np
from scipy import stats
```

```python
# Read CSV with explicit date parsing
m1 = pd.read_csv("csv1/AfterTest1.csv", index_col=8, parse_dates=[8], dayfirst=True)

# Use iloc for row slicing to avoid future warnings
m1 = m1.iloc[2107:2900]
```

```python
print(m1.columns)
```

```python
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline
fig = plt.figure()
pylab.rcParams['figure.figsize'] = (15, 10)  
plt.plot(m1[['cash']],'r',label='Proposed Model1',linewidth=0.75)
plt.title('NIFTY')
plt.ylabel('% Accumulated Return')
plt.xlabel('Testing Period (days)')
plt.legend()
plt.grid(True,color='k')
plt.show()
fig.savefig('Areturn_nifty.png')
```
