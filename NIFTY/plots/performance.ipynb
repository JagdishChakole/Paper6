```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn
import numpy as np
from scipy import stats
```

```python
# Read CSV with explicit date parsing
m1 = pd.read_csv("asr1.csv", index_col=7, parse_dates=[7], dayfirst=True)

# Use iloc for row slicing to avoid future warnings
m1 = m1.iloc[2107:2844]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Calculate Daily Return
m1["Daily Return"] = m1["cash"].pct_change()

# Calculate Cumulative Return
m1["Cumulative Return"] = (1 + m1["Daily Return"]).cumprod()

# Calculate Drawdown
m1["Drawdown"] = m1["cash"] / m1["cash"].cummax() - 1

# Compute Performance Metrics
max_drawdown = m1["Drawdown"].min()
risk_free_rate_annual = 0.03  # Example: 3% risk-free rate (annual)
risk_free_rate_daily = risk_free_rate_annual / 252  # Convert to daily
sharpe_ratio = (m1["Daily Return"].mean() - risk_free_rate_daily) / m1["Daily Return"].std() * np.sqrt(252)


# Plot Cash Balance & Performance
plt.figure(figsize=(12, 5))
plt.plot(m1["cash"], label="Cash Balance", color="blue")
plt.plot(m1["Cumulative Return"] * m1["cash"].iloc[0], label="Cumulative Return", linestyle="dashed", color="green")
plt.xlabel("Time")
plt.ylabel("Cash Balance")
plt.legend()
plt.title("Trading Strategy Performance")
plt.show()

# Print Performance Metrics
print(f"Total Return: {m1['Cumulative Return'].iloc[-1] - 1:.2%}")
print(f"Max Drawdown: {max_drawdown:.2%}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")

# Compute additional performance metrics
average_daily_return = m1["Daily Return"].mean()
std_dev_daily_return = m1["Daily Return"].std()

# Print additional performance metrics
print(f"Average Daily Return: {average_daily_return * 100:.2f}%")
print(f"Standard Deviation of Daily Return: {std_dev_daily_return * 100:.2f}%")
# Number of winning days (profitable days)
winning_days = (m1["diff"] > 0).sum()

# Total number of days (trades)
total_days = len(m1)

# Win Rate calculation
win_rate = (winning_days / total_days) * 100

# Print the Win Rate
print(f"Win Rate: {win_rate:.2f}%")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute Buy-and-Hold Daily Returns using 'Close' prices
m5["BuyHold Daily Return"] = m5["Close"].pct_change()

# Compute Buy-and-Hold Cumulative Returns
m5["BuyHold Cumulative Return"] = (1 + m5["BuyHold Daily Return"]).cumprod()

# Compute Buy-and-Hold Drawdown
m5["BuyHold Drawdown"] = m5["Close"] / m5["Close"].cummax() - 1
max_drawdown_buyhold = m5["BuyHold Drawdown"].min()

# Compute Buy-and-Hold Sharpe Ratio
risk_free_rate_annual = 0.03  # Example: 3% annual risk-free rate
risk_free_rate_daily = risk_free_rate_annual / 252  # Convert to daily rate

sharpe_ratio_buyhold = (m5["BuyHold Daily Return"].mean() - risk_free_rate_daily) / m5["BuyHold Daily Return"].std() * np.sqrt(252)

# Plot Buy-and-Hold Performance
plt.figure(figsize=(12, 5))
plt.plot(m5["Close"], label="Stock Price (Close)", color="blue")
plt.plot(m5["BuyHold Cumulative Return"] * m5["Close"].iloc[0], label="Buy-and-Hold Cumulative Return", linestyle="dashed", color="green")
plt.xlabel("Time")
plt.ylabel("Stock Price / Return")
plt.legend()
plt.title("Buy-and-Hold Strategy Performance")
plt.show()

# Compute Additional Performance Metrics
average_daily_return_buyhold = m5["BuyHold Daily Return"].mean()
std_dev_daily_return_buyhold = m5["BuyHold Daily Return"].std()

# Compute Buy-and-Hold Win Rate
winning_days_buyhold = (m5["BuyHold Daily Return"] > 0).sum()
total_days_buyhold = len(m5)
win_rate_buyhold = (winning_days_buyhold / total_days_buyhold) * 100

# Print Buy-and-Hold Performance Metrics
print(f"Buy-and-Hold Total Return: {m5['BuyHold Cumulative Return'].iloc[-1] - 1:.2%}")
print(f"Buy-and-Hold Max Drawdown: {max_drawdown_buyhold:.2%}")
print(f"Buy-and-Hold Sharpe Ratio: {sharpe_ratio_buyhold:.2f}")
print(f"Buy-and-Hold Average Daily Return: {average_daily_return_buyhold * 100:.2f}%")
print(f"Buy-and-Hold Std Dev of Daily Return: {std_dev_daily_return_buyhold * 100:.2f}%")
print(f"Buy-and-Hold Win Rate: {win_rate_buyhold:.2f}%")
```
