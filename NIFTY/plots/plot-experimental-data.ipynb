```python
import numpy as np
import datetime
import pandas as pd
import random
```

```python
 X=pd.read_csv('nifty.csv')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'date' column to datetime (your column name is lowercase)
X['date'] = pd.to_datetime(X['date'], dayfirst=True)

# Set 'date' as index
X.set_index('date', inplace=True)

# Define the date ranges using the original index positions
train_end_date = X.index[2207]
test_end_date = X.index[2944]

# Plot settings
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 12

fig = plt.figure()
ax = fig.add_subplot(111)

# Plot training data
X.loc[:train_end_date, 'Close'].plot(ax=ax, label='Training Data', color='blue', linewidth=1.5)

# Plot testing data
X.loc[train_end_date + pd.Timedelta(days=1):test_end_date, 'Close'].plot(ax=ax, label='Testing Data', color='orange', linewidth=1.5)

# Add vertical line at split
plt.axvline(x=train_end_date, color='gray', linestyle='--', linewidth=1.5, label='Train/Test Split')

# Add labels and title
plt.title('NIFTY Index Stock Close Price', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Close Price (INR)', fontsize=14)

# Add grid and legend
plt.grid(True, color='k', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)
fig.autofmt_xdate()

# Show plot
plt.tight_layout()
plt.show()

# Save high-res image
fig.savefig('nifty_data.png', dpi=500, bbox_inches='tight')
```
