```python
import yfinance as yf
import pandas as pd

# Define stock symbol for NSE
ticker = "RELIANCE.NS"

# Fetch minute-wise data for the last 1 day
df = yf.download(ticker, period="8d", interval="5m")

# Check if the index is already timezone-aware
if not df.index.tz:
    # If not, localize to UTC
    df.index = df.index.tz_localize('UTC')

# Convert the index to Indian Standard Time (IST)
df.index = df.index.tz_convert('Asia/Kolkata')
```

```python
import yfinance as yf
import pandas as pd

# Define stock symbol for NSE
ticker = "RELIANCE.NS"

# Specify start and end dates
start_date = "2024-03-29"  # YYYY-MM-DD format
end_date = "2024-03-30"

# Fetch minute-wise data
df = yf.download(ticker, start=start_date, end=end_date, interval="1m")

# Convert the index to Indian Standard Time (IST)
if not df.index.tz:
    df.index = df.index.tz_localize('UTC')
df.index = df.index.tz_convert('Asia/Kolkata')  # ✅ Fixed closing quote

# Display data
print(df.head())
```
