```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('nifty_clean_after_XAI.csv', index_col=0, low_memory=False)
```

```python
df.columns
```

```python
# Specify the columns to clean (replace with actual column names)
columns_to_clean = ['weekday', 'month',
       'SMA_10', 'SMA_50', 'SMA_100', 'EMA_5', 'EMA_10', 'EMA_20',
       'MACD_12_26_9', 'MACDh_12_26_9', 'MACDs_12_26_9', 'KAMA', 'TMA_20',
       'RSI_7', 'RSI_14', '%D', 'ROC_5', 'ROC_10', 'ROC_20', 'CCI_14',
       'CCI_50', 'MOM_10', 'UO', 'KST', 'KST_Signal', 'BB_Upper', 'BB_Middle',
       'BB_Lower', 'BB_PercentB', 'HL_Range', 'HL_EMA', 'CVI', 'STDDEV_10',
       'STDDEV_20', 'Log_Returns', 'HV_10', 'HV_20', 'GK_Vol_10', 'GK_Vol_20',
       'OBV', 'MFM', 'MFV', 'AD', 'CMF_20', 'FI_13', 'VROC_10', 'PVT']  # Update column names
```

```python
# Replace Inf and -Inf with NaN **starting from row 100**
df.loc[100:, columns_to_clean] = df.loc[100:, columns_to_clean].replace([np.inf, -np.inf], np.nan)
```

```python
# Handle NaN values (e.g., filling with the column mean **from row 100**)
df.loc[100:, columns_to_clean] = df.loc[100:, columns_to_clean].fillna(df.loc[100:, columns_to_clean].mean())
```

```python
df.to_csv('input2lstm.c', index=False)
```
