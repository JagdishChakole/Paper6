```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('nifty_clean_after_XAI.csv', index_col=0, low_memory=False)
```

```python
df.head()
```

```python
data_subset = df.iloc[100:].drop(columns=['date']).copy()
```

```python
data_subset.columns
```

```python
print(data_subset.isna().sum())
```

```python
print(data_subset.isna().any().any())  # True if any NaN exists, False otherwise
```

```python
nan_rows = data_subset[data_subset.isna().any(axis=1)]
print(nan_rows)
```

```python
data_subset = data_subset.fillna(data_subset.mean())  # Replace NaN with column means
```

```python
print(np.isinf(data_subset).sum())  # Count Inf values in each column
print(np.isinf(data_subset).any().any())  # True if any Inf exists
```

```python
for col in data_subset.select_dtypes(include=[np.number]):  # Select numeric columns
    max_value = data_subset.loc[data_subset[col] != np.inf, col].max()  # Max excluding inf
    data_subset[col] = data_subset[col].replace([np.inf, -np.inf], max_value)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Apply K-Means on the selected rows
kmeans = KMeans(n_clusters=6, random_state=42)
data_subset['Cluster'] = kmeans.fit_predict(data_subset[['weekday',
       'SMA_10', 'SMA_50', 'SMA_100', 'EMA_5', 'EMA_10', 'EMA_20',
       'MACD_12_26_9', 'MACDh_12_26_9', 'MACDs_12_26_9', 'KAMA', 'TMA_20',
       'RSI_7', 'RSI_14', '%D', 'ROC_5', 'ROC_10', 'ROC_20', 'CCI_14',
       'CCI_50', 'MOM_10', 'UO', 'KST', 'KST_Signal', 'BB_Upper', 'BB_Middle',
       'BB_Lower', 'BB_PercentB', 'HL_Range', 'HL_EMA', 'CVI', 'STDDEV_10',
       'STDDEV_20', 'Log_Returns', 'HV_10', 'HV_20', 'GK_Vol_10', 'GK_Vol_20',
       'OBV', 'MFM', 'MFV', 'AD', 'CMF_20', 'FI_13', 'VROC_10', 'PVT']])
```

```python
# Store the clustered values back in the original DataFrame
df.loc[100:, 'Cluster'] = data_subset['Cluster'].values
```

```python
#df.index = pd.to_datetime(df.index)
df = df.reset_index(drop=True)
df.to_csv('state_after_kmeans.csv', index=True) 
```
